```python
# 1. Prepare data for training
# Data Prep: select appropriate features, address missing values, find/handle outliers, look at final correlations
```

```python
#removing 'country' because of categorical dtype
#removing 'year' because the problem doesn't focus on year
#removing 'Standard deviation of ladder by country-year' and 'Standard deviation/Mean of ladder by country-year' because of connection to 'year'
#removing 'Democratic Quality' and 'Delivery Quality' because of negative values
df = df.drop(columns='country')
df = df.drop(columns='year')
df = df.drop(columns='Democratic Quality')
df = df.drop(columns='Delivery Quality')
df = df.drop(columns='Standard deviation of ladder by country-year')
df = df.drop(columns='Standard deviation/Mean of ladder by country-year')
```

```python
#going to remove 'GINI index (World Bank estimate)' due to large number of null values
#removing 'GINI index (World Bank estimate), average 2000-15' and 'gini of household 
#income reported in Gallup, by wp5-year' due to their connection to 'year'

df = df.drop(columns='GINI index (World Bank estimate)')
df = df.drop(columns='GINI index (World Bank estimate), average 2000-15')
df = df.drop(columns='gini of household income reported in Gallup, by wp5-year')
```

```python
#handle null values by replacing with the mean of the column

df['Log GDP per capita'].fillna(value=df['Log GDP per capita'].mean(), inplace=True)
df['Social support'].fillna(value=df['Social support'].mean(), inplace=True)
df['Healthy life expectancy at birth'].fillna(value=df['Healthy life expectancy at birth'].mean(), inplace=True)
df['Freedom to make life choices'].fillna(value=df['Freedom to make life choices'].mean(), inplace=True)
df['Generosity'].fillna(value=df['Generosity'].mean(), inplace=True)
df['Perceptions of corruption'].fillna(value=df['Perceptions of corruption'].mean(), inplace=True)
df['Positive affect'].fillna(value=df['Positive affect'].mean(), inplace=True)
df['Negative affect'].fillna(value=df['Negative affect'].mean(), inplace=True)
df['Confidence in national government'].fillna(value=df['Confidence in national government'].mean(), inplace=True)
```

```python
#double check there are no null values
print(np.sum(df.isnull(), axis=0))
```

```python
#handle outliers using SciPy and winsorize(top/bottom 1%)

df['Life Ladder'] = stats.mstats.winsorize(df['Life Ladder'], limits=[0.01, 0.01])
df['Log GDP per capita'] = stats.mstats.winsorize(df['Log GDP per capita'], limits=[0.01, 0.01])
df['Social support'] = stats.mstats.winsorize(df['Social support'], limits=[0.01, 0.01])
df['Healthy life expectancy at birth'] = stats.mstats.winsorize(df['Healthy life expectancy at birth'], limits=[0.01, 0.01])
df['Freedom to make life choices'] = stats.mstats.winsorize(df['Freedom to make life choices'], limits=[0.01, 0.01])
df['Generosity'] = stats.mstats.winsorize(df['Generosity'], limits=[0.01, 0.01])
df['Perceptions of corruption'] = stats.mstats.winsorize(df['Perceptions of corruption'], limits=[0.01, 0.01])
df['Positive affect'] = stats.mstats.winsorize(df['Positive affect'], limits=[0.01, 0.01])
df['Negative affect'] = stats.mstats.winsorize(df['Negative affect'], limits=[0.01, 0.01])
df['Confidence in national government'] = stats.mstats.winsorize(df['Confidence in national government'], limits=[0.01, 0.01])
```

```python
#look at correlation between features and label
print(df.corr()['Life Ladder'])
```

```python
#drop the label ('Life Ladder') from the df of features
df_features = df.drop(columns='Life Ladder')
df_features.head()
```

```python
#make 'Life Ladder' label into a binary variable (makes this a classification problem)
#if value is below 5 -> 0 
#if value is above 5 -> 1 

condition_below_5 = ~(df['Life Ladder'] < 5.0000000)
condition_above_5 = ~(df['Life Ladder'] > 5.0000000)

df['Life Ladder'] = np.where(condition_below_5, 0 , df['Life Ladder'])
df['Life Ladder'] = np.where(condition_above_5, 1 , df['Life Ladder'])

df['Life Ladder'].unique()
```

```python
# split data into testing/training sets to avoid overfitting/bias
y = df['Life Ladder']
X = df_features

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1234)
```

```python
# Fit model to training data and evaluate the 3 models
```

```python
# Logisitic regression (initial implementation)
logr_model = LogisticRegression(C=1)
logr_model.fit(X_train, y_train)

logr_prob_predict = logr_model.predict_proba(X_test)
logr_loss = log_loss(y_test, logr_prob_predict)
    
logr_label_predict = logr_model.predict(X_test)
logr_acc = accuracy_score(y_test, logr_label_predict)

print('Initial accuracy: ' + str(logr_acc))
print('Initial loss: ' + str(logr_loss))
```

```python
# optimize linear regression by testing out different values of C
# calculate accuracy and loss (evaluation metrics) as well

c_list = range(1, 10)
logr_loss_vals = []
logr_accuracy_vals = []

for c in c_list:
    log_model = LogisticRegression(C=c)
    log_model.fit(X_train, y_train)

    log_prob_predict_o = log_model.predict_proba(X_test)
    log_loss_o = log_loss(y_test, log_prob_predict_o)
    
    log_label_predict_o = log_model.predict(X_test)
    log_acc_o = accuracy_score(y_test, log_label_predict_o)

    logr_loss_vals.append(log_loss_o)
    logr_accuracy_vals.append(log_acc_o)

acc_index = logr_accuracy_vals.index(max(logr_accuracy_vals)) + 1
loss_index = logr_loss_vals.index(min(logr_loss_vals)) + 1

print('Max accuracy: ' + str(max(logr_accuracy_vals)) + ' at C=' + str(acc_index))
print('Min loss: ' + str(min(logr_loss_vals)) + ' at C=' + str(loss_index))
```

```python
#visualize logistic regression loss
sns.lineplot(x=c_list, y=logr_loss_vals)
plt.xlabel('Hyperparameter C')
plt.ylabel('Log Loss')
```

```python
#visualize logistic regression accuracy
sns.lineplot(x=c_list, y=logr_accuracy_vals)
plt.xlabel('Hyperparameter C')
plt.ylabel('Accuracy')
```

```python
# Best logistic regression when:
# - C=9 (highest accuracy), C=4 (lowest loss)

# Analysis:
# C=4 means the lowest accuracy for lower loss
# C=9 means higher accuracy for the highest loss (This could be an indication of overfitting.)

# When analyzing the metric graphs above, C=4 has a relatively high accuracy compared to C=3 and C=5. 
# Since C=4 also has the lowest loss, C=4 is the best hyperparameter!
print('Final logistic regression accuracy: ' + str(logr_accuracy_vals[4]))
print('Final logistic regression loss: ' + str(logr_loss_vals[4]))

```

```python
# decision tree (initial implementation)
dt_model = DecisionTreeClassifier(max_depth=4, min_samples_leaf=1)    
dt_model.fit(X_train, y_train)
dt_label_predict = dt_model.predict(X_test)
dt_acc_val = accuracy_score(y_test, dt_label_predict)

dt_prob_predict = dt_model.predict_proba(X_test)
dt_loss_val = log_loss(y_test, dt_prob_predict)

print('Initial accuracy: ' + str(dt_acc_val))
print('Initial loss: ' + str(dt_loss_val))
```

```python
# optimize decision tree using a grid search
depth_vals = [4, 8, 16, 32]
leaf_vals = [25, 50, 100, 200]
hyp_grid={'max_depth':depth_vals, 'min_samples_leaf':leaf_vals}

dt_model_o = DecisionTreeClassifier()
dt_grid = GridSearchCV(dt_model_o, hyp_grid, cv=5)
dt_grid_search = dt_grid.fit(X_train, y_train)

print('Optimal hyperparameters: {0}'.format(dt_grid_search.best_params_))
print('Best accuracy score: {0}'.format(dt_grid_search.best_score_))
```

```python
# Best decision tree when:
# - max_depth=4, min_samples_leaf=50

#Implement decision tree again with best hyperparameters
dt_model = DecisionTreeClassifier(max_depth=4, min_samples_leaf=50)    
dt_model.fit(X_train, y_train)
dt_label_predict = dt_model.predict(X_test)
dt_acc_val = accuracy_score(y_test, dt_label_predict)

dt_prob_predict = dt_model.predict_proba(X_test)
dt_loss_val = log_loss(y_test, dt_prob_predict)

print('Final decision tree accuracy: ' + str(dt_acc_val))
print('Final decision tree loss: ' + str(dt_loss_val))
```

```python
# neural network (initial implementation)
def neuralN(l_rate, epochs, val_split):
    nn_model = keras.Sequential()

    input_shape = X_train.shape[1]
    input_layer = keras.layers.InputLayer(input_shape)
    nn_model.add(input_layer)

    h_layer_test = keras.layers.Dense(units=128, activation='relu')
    nn_model.add(h_layer_test)

    h_layer1 = keras.layers.Dense(units=64, activation='relu')
    nn_model.add(h_layer1)

    h_layer2 = keras.layers.Dense(units=32, activation='relu')
    nn_model.add(h_layer2)

    h_layer3 = keras.layers.Dense(units=16, activation='relu')
    nn_model.add(h_layer3)

    output_layer = keras.layers.Dense(units=1, activation='relu')
    nn_model.add(output_layer)

    sgd_optimizer = keras.optimizers.SGD(learning_rate=l_rate)
    loss_fn = keras.losses.BinaryCrossentropy(from_logits=False)
    nn_model.compile(optimizer=sgd_optimizer, loss=loss_fn, metrics=['accuracy'])

    history = nn_model.fit(X_train, y_train, epochs=epochs, verbose=0, validation_split=val_split)

    nn_loss_val, nn_accuracy_val = nn_model.evaluate(X_test, y_test)

    print('Initial accuracy: ' + str(nn_accuracy_val))
    print('Initial loss: ' + str(nn_loss_val))
```

```python
#optimize neural network by testing out different values for learning_rate, epochs, and validation_split
neuralN(0.1, 50, 0.2)
neuralN(0.1, 100, 0.2)
neuralN(0.1, 100, 0.4)
neuralN(0.2, 100, 0.4)
neuralN(0.4, 150, 0.6)
neuralN(0.5, 200, 0.6)
```
